```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
import numpy as np

generalData = pd.read_csv('ReporteDatosGenerales.csv')
listadoOperacion = pd.read_csv('ReporteListadoOperacion.csv')
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
print(len(generalData))
print(len(listadoOperacion))
print(display(generalData.head()))
print(display(listadoOperacion.head()))
```

```python
#Clean Data, Lets Start with renaming de Columns

print(generalData['asesorDataTextBox'])
print(generalData['textBox13'])
```

```python
#Create Data Frames with new columns names
l_o = pd.DataFrame()
g_d = pd.DataFrame()

#Lets create clean General Data DataFrame

g_d['nombre'] = generalData['asesorDataTextBox']
#g_d.set_index('name', inplace = True)
g_d['genero'] = generalData['textBox13']
g_d['rfc'] = generalData['textBox6']
g_d['actividad'] = generalData['textBox28']
g_d['colonia'] = generalData['textBox34']
print(display(g_d))
```

```python
#Lets create clean listado Operacion Dataframe

l_o['nombre'] = listadoOperacion['asesorDataTextBox']
l_o['fecha nacimiento'] = listadoOperacion['textBox33']
l_o['producto'] = listadoOperacion['textBox18']
l_o['importe'] = listadoOperacion['textBox38']
l_o['importe a pagar'] = listadoOperacion['textBox16']
l_o['# de fallas'] = listadoOperacion['textBox15']
l_o['estatus'] = listadoOperacion['textBox24']
l_o['cuotas pagadas'] = listadoOperacion['textBox28']
l_o['pagos anticipados'] = listadoOperacion['textBox32']
l_o['fecha inicio'] = listadoOperacion['textBox36']
l_o['fecha fin'] = listadoOperacion['textBox2']

print(display(l_o.head()))
```

```python
print(g_d.dtypes)
l_o.importe = l_o['importe'].replace('[$]', '', regex = True)
l_o.importe = l_o['importe'].replace('[,]', '', regex = True)
l_o.importe = l_o['importe'].str[:-3]
print(l_o.importe)

convert_dict = {
    'importe': int, 
    'importe a pagar': int, 
    '# de fallas': int, 
    'cuotas pagadas': int, 
    'pagos anticipados': int}
```

```python
l_o['importe a pagar'] = l_o['importe a pagar'].replace('[$]', '', regex = True)
l_o['importe a pagar'] = l_o['importe a pagar'].replace('[,]', '', regex = True)
l_o['importe a pagar'] = l_o['importe a pagar'].str[:-3]
print(l_o['importe a pagar'])
```

```python

l_o['importe'] = l_o['importe'].apply(pd.to_numeric)
l_o['importe a pagar'] = l_o['importe a pagar'].apply(pd.to_numeric)
l_o['# de fallas'] = l_o['# de fallas'].apply(pd.to_numeric)

print(l_o.dtypes)
```

```python
print(display(round((l_o.describe()))),4)
print(round(l_o['# de fallas'].median()),2)
```

```python
#Merge both data sets 

dataset = pd.merge(g_d, l_o, how = 'inner',on='nombre')
```

```python
print(display(dataset.head()))
```

```python
rfclam = dataset['rfc'].apply(
    lambda x: 'No' if x == 'XAXX010101000' else 'Yes')

print(rfclam)
dataset['rfc'] = rfclam
```

```python
#Modifiquemos la edad
ano = dataset['fecha nacimiento'].str[-4:]
dataset['fecha nacimiento'] = ano
print(display(dataset.head()))
```

```python
dataset['fecha nacimiento'] = dataset['fecha nacimiento'].apply(pd.to_numeric)
dataset['edad'] = dataset['fecha nacimiento'].apply(lambda x: 2024 - x)
```
